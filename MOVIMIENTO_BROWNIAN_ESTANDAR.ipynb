# trayectoria individual

```python


np.random.seed(0)

T = 0.25
n = 252
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W1 = np.cumsum(increments)

plt.figure(figsize=(10,5))
plt.plot(W1)
plt.title("Trayectoria de movimiento browniano con 0.25 t")
plt.xlabel("Tiempo")
plt.ylabel("W(t)")
plt.show()
```
