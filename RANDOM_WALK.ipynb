# simulacion de un random walk 

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
np.random.seed (42)

n_steps= 200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()
```

#  Muchas simulaciones  aparece la distribucion normal 

```python

np.random.seed(1)

n_sim = 5000
n_steps = 200

final_positions = []

for _ in range(n_sim):
    steps = np.random.choice([-1, 1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=50)
plt.title("Distribucin de posiciones finales")
plt.show()


#Observacin:

 #La distribucin se vuelve aproximadamente no
```

# La incertidumbre  crece con el tiempo 

```python

np.random.seed(0)

n_sim = 2000
times = [10, 50, 100, 200, 400]

for t in times:
    finals = []
    for _ in range(n_sim):
        steps = np.random.choice([-1, 1], size=t)
        finals.append(np.sum(steps))
    print(f"t={t}, var{np.var(finals):.2f}")
```

# Primera aproximacion a movimiento browniano 

```python
np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()
```

```python
np.random.seed(42)
# copio el mismo codigo  para el randome walk 
n_sim = 200
n_steps = 200

paths = np.zeros((n_sim, n_steps))

for i in range(n_sim):
    steps = np.random.choice([-1, 1], size=n_steps)
    paths[i] = 100 + np.cumsum(steps)

mean_path = np.mean(paths, axis=0)
centered_paths = paths - mean_path


# ahora lo grafica 
plt.figure(figsize=(10,5))
for i in range(50):
    plt.plot(centered_paths[i], alpha=0.4)

plt.axhline(0, linestyle="--")
plt.title("Random Walks centrados (media cero)")
plt.xlabel("Tiempo")
plt.ylabel("Desviacin respecto al promedio")
plt.show()

```
